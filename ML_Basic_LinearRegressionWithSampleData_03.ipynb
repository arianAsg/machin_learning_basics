# Importing librarys

we will use  stats from scipy library for statistical calculation

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
```

# Linear Regression

#### First step: generating sample data

```python
x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]

plt.scatter(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.show() 
```

#### Second step: use stats.linregrees() to calculate important parameter of the linear regression

```python
slope, intercept, r, p, std_err = stats.linregress(x, y) 
```

```python
new_x=np.tile(slope,(13))
new_y=x*new_x+np.tile(intercept,(13))
```

```python
plt.scatter(x, y)
plt.plot(x, new_y)
plt.show()
```
